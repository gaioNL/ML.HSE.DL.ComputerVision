## Item-Item Collaborative Filtering Assigment

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
input_xls = pd.ExcelFile('Assignment 5.xls')
```

```python
ratings_df = pd.read_excel(input_xls,'Ratings')
normRatings_df = pd.read_excel(input_xls,'NormRatings')
results_df = pd.read_excel(input_xls,'Matrix')
```

### Cosine Simularity

```python
cos_raw_df = pd.DataFrame(
            cosine_similarity(ratings_df.iloc[:-1,1:-1].fillna(0).T),
            index = results_df.index, columns = results_df.columns)
```

```python
cos_raw_df
```

```python
cos_norm_df = pd.DataFrame(
            cosine_similarity(normRatings_df.iloc[:-1,1:].fillna(0).T),
            index = results_df.index, columns = results_df.columns)
```

```python
cos_norm_df
```

### Top 5 for User 5277

Provide the top 5 recommended movies for user 5277, using an average of the user’s ratings weighted by similarity to each candidate movie.

You do not need to exclude movies they have rated. 

Consider all movies with nonnegative similarities (do not limit neighborhood size).

```python
def get_rec_raw(w,u,r,raw=True) #similarities(weights),user,ratings
    user_ratings = r.loc[r.User==u]
    user_rated_movies= user_ratings.columns[pd.notnull(user_ratings).values[0]]
    i_movie = w.name
    j_movies = user_rated_movies.tolist()
    if raw:
        j_movies.remove('Mean')
    j_movies.remove('User')        
    wij = w.loc[j_movies].values
    ratings=r.loc[r.User==u,j_movies]
    
    return np.dot(wij,ratings.values[0])/np.sum(wij)
    
```

```python
raw_rec= cos_raw_df.apply(get_rec_raw,axis=1,args=(5277,ratings_df,True))
```

```python
raw_rec.sort_values(ascending=False)[0:5]
```

```python
cos_norm_df = cos_norm_df.applymap(lambda x:max(0,x))
```

```python
norm_rec = cos_norm_df.apply(get_rec_raw,axis=1,args=(5277,ratings_df,True))
```

```python
norm_rec.sort_values(ascending=False)[0:5]
```
