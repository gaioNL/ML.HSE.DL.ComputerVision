```python
def top_n_neighbors(user_movie_df,user,n=5):
    uu_corr_df = user_movie_df.corr()
    return uu_corr_df.loc[user].sort_values(ascending=False ).iloc[1:n+1] 
```

```python
def predict_item(user_movie_df,user,n=5,l=12):
    top_nn_df = top_n_neighbors(user_movie_df,user,n)#top n correlated users
    ratings_top_nn_df = uu_df.loc[:,top_nn_df.index] #movie ratings from top n users
    den_pred_df = ratings_top_nn_df.notnull().mul(top_nn_df).sum(axis=1) # denominator prediction formula
    ratings_top_nn_df.fillna(0) #movie ratings from top n users
    num_pred_df = ratings_top_nn_df.mul(top_nn_df).sum(axis=1) # numerator of prediction formula
    pred_items_df = num_pred_df.div(den_pred_df)
    return  pred_items_df.sort_values(ascending=False ).iloc[0:l]
```

```python
predict_item(uu_df,3867,5,3)
```

```python
def predict_item_normalized(user_movie_df,user,n=5,l=12):
    top_nn_df = top_n_neighbors(user_movie_df,user,n)#top n correlated users
    ratings_top_nn_df = uu_df.loc[:,top_nn_df.index] #movie ratings from top n users
    
    mean_user =  user_movie_df[user].mean()
    
    den_pred_df = ratings_top_nn_df.notnull().mul(top_nn_df).sum(axis=1) # denominator prediction formula
    
    mean_top_nn_df = ratings_top_nn_df.mean(axis=0)
    
    ratings_top_nn_df.fillna(0) #movie ratings from top n users
    num_pred_df = (ratings_top_nn_df-mean_top_nn_df).mul(top_nn_df).sum(axis=1) # numerator of prediction formula
    pred_items_df = mean_user +  num_pred_df.div(den_pred_df)
    return  pred_items_df.sort_values(ascending=False ).iloc[0:l]
```

For example, if the user ID was 3712, the correct submission would be:

Top Movie: 641 Prediction: 5.900

2nd Movie: 603 Prediction: 5.546

3rd Movie: 105 Prediction: 5.501

```python
predict_item_normalized(uu_df,3712,5,3)
```

```python
predict_item_normalized(uu_df,3867,5,3)
```

```python
predict_item_normalized(uu_df,89,5,3)
```
