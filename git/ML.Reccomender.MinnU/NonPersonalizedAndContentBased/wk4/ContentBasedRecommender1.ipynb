```python
import pandas as pd
import numpy as np
```

```python
reviews_df = pd.read_csv('reviews.csv')
```

```python
reviews_df.head()
```

```python
reviews_df = reviews_df.fillna(0)
```

```python
reviews_df.head()
```

### Part 1

```python
profiles_dot_df = np.dot(reviews_df.loc[:,'baseball':'family'].T,reviews_df.loc[:,'User 1':'User 2' ])
```

```python
profiles_df= pd.DataFrame(data=profiles_dot_df.T,columns = reviews_df.columns[1:-3]  )
```

```python
profiles_df['user'] = ['User 1','User 2']
```

```python
profiles_df
```

#### 1) Which document does the simple profile predict user 1 will like best?
#### 2) What score does that prediction get?

```python
docs_users_rating = np.dot(reviews_df.loc[:,'baseball':'family'],profiles_df.loc[:,'baseball':'family'].T)
```

```python
docs_users_rating_df = pd.DataFrame(data = docs_users_rating, columns = ['User 1','User 2' ])
```

```python
docs_users_rating_df['doc'] = reviews_df.iloc[:,0]
```

```python
docs_users_rating_df.sort_values(by = 'User 1', ascending = False)
```

#### How many documents does the model predict User 2 will dislike (prediction score that is negative)?

```python
docs_users_rating_df[docs_users_rating_df['User 2'] < 0].count()
```

### Part 2

```python
p2_reviews_df = reviews_df.copy()
```

```python
np.sqrt(p2_reviews_df['num-attr']).head()
```

```python
p2_reviews_df.loc[:,'baseball':'family']= p2_reviews_df.loc[:,'baseball':'family'].div(np.sqrt(p2_reviews_df['num-attr']),axis = 0)
```

```python
p2_reviews_df
```

```python
profiles_dot_norm_df = np.dot(p2_reviews_df.loc[:,'baseball':'family'].T,p2_reviews_df.loc[:,'User 1':'User 2' ])
```

```python
profiles_norm_df= pd.DataFrame(data=profiles_dot_norm_df.T,columns = p2_reviews_df.columns[1:-3]  )
```

```python
profiles_norm_df['user'] = ['User 1','User 2']
```

```python
profiles_norm_df
```

```python
docs_users_rating_norm = np.dot(p2_reviews_df.loc[:,'baseball':'family'],profiles_norm_df.loc[:,'baseball':'family'].T)
```

```python
docs_users_rating_norm_df = pd.DataFrame(data = docs_users_rating_norm, columns = ['User 1','User 2' ])
```

```python
docs_users_rating_norm_df['doc'] = p2_reviews_df.iloc[:,0]
```

```python
docs_users_rating_norm_df.sort_values(by = 'User 1', ascending = False)
```
